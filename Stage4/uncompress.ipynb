```python
import os
import subprocess

# 定义源目录和目标目录
source_dir = "/N/project/Zli_lab/Data/Observations/stage4/"
target_dir = "/N/project/Zli_lab/Data/Observations/stage4_uncompress/"

# 确保目标目录存在
os.makedirs(target_dir, exist_ok=True)

# 遍历源目录中的所有子目录
for subdir in sorted(os.listdir(source_dir)):
    subdir_path = os.path.join(source_dir, subdir)

    # 确保它是一个目录
    if os.path.isdir(subdir_path):
        print(f"Processing directory: {subdir}")

        # 遍历子目录中的 .Z 文件
        for file in os.listdir(subdir_path):
            if file.endswith(".Z"):
                file_path = os.path.join(subdir_path, file)
                output_file = os.path.join(target_dir, file[:-2])  # 去掉 `.Z` 扩展名

                # 解压文件
                try:
                    subprocess.run(["uncompress", "-c", file_path], stdout=open(output_file, "wb"), check=True)
                    print(f"Decompressed: {file} -> {output_file}")
                except subprocess.CalledProcessError as e:
                    print(f"Error decompressing {file}: {e}")
```

```python
import os
import subprocess

# 定义源目录和目标目录
source_dir = "/N/project/Zli_lab/Data/Observations/stage4/"
target_dir = "/N/project/Zli_lab/Data/Observations/stage4_uncompress/"

# 确保目标目录存在
os.makedirs(target_dir, exist_ok=True)

# 遍历源目录中的所有子目录
for subdir in sorted(os.listdir(source_dir)):
    subdir_path = os.path.join(source_dir, subdir)

    # 确保它是一个目录
    if os.path.isdir(subdir_path):
        print(f"Processing directory: {subdir}")

        # 遍历子目录中的 .Z 文件
        for file in os.listdir(subdir_path):
            if file.endswith(".Z"):
                file_path = os.path.join(subdir_path, file)
                output_file = os.path.join(target_dir, file[:-2])  # 去掉 `.Z` 扩展名

                # 解压文件
                try:
                    with open(output_file, "wb") as out_f:
                        subprocess.run(["gzip", "-d", "-c", file_path], stdout=out_f, check=True)
                    print(f"Decompressed: {file} -> {output_file}")
                except subprocess.CalledProcessError as e:
                    print(f"Error decompressing {file}: {e}")
```
