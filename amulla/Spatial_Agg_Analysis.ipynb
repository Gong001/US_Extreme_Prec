```python
avg_decade_1979_1988 = jja_file[:10].mean(dim="year",skipna=True)
avg_decade_1989_1998 = jja_file[10:20].mean(dim="year",skipna=True)
avg_decade_3 = jja_file[20:30].mean(dim="year",skipna=True)
avg_decade_4 = jja_file[30:40].mean(dim="year",skipna=True)

num_rows = 2
num_cols = 4
regions = ['Northeast','Midwest','Southeast','Northern Great Plains','Southern Great Plains','Southwest','Northwest']
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(16, 8))  # Adjust figsize as needed

M = [avg_decade_1979_1988,avg_decade_1989_1998,avg_decade_3,avg_decade_4]
LABEL = ['1979-','1989-','1999-','2009-']

for i in range(num_rows):
    for j in range(num_cols):
        index = i * num_cols + j
        if index < 7:
            for k in range(4):
                axes[i, j].plot(M[k][:, index]*1000,label=LABEL[k])
            
            axes[i, j].set_title(f'{regions[index]}')
            axes[i, j].set_xlabel('Hour')
            if (i==0 and j==0) or (i==1 and j==0):
                axes[i, j].set_ylabel('JJA Extreme Precipitation (mm)')
        else:
            axes[i, j].axis('off')  # Hide the empty subplot
        axes[i, j].legend()
plt.tight_layout()
plt.show()
```
