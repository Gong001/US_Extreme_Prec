```python
import numpy as np
import xarray as xr

def conservative_downscale_04_to_01(ds):
    """
    将 0.04° 分辨率数据降尺度到 0.1° 分辨率，使用区域平均（conservative downscaling）
    """
    # 确保 south_north 和 west_east 维度是偶数
    sn_size, we_size = ds.dims['south_north'], ds.dims['west_east']
    
    if sn_size % 2 != 0:
        ds = ds.isel(south_north=slice(0, sn_size - 1))
    if we_size % 2 != 0:
        ds = ds.isel(west_east=slice(0, we_size - 1))

    # 计算缩放因子
    lat_factor = int(0.1 / 0.04)  # 2
    lon_factor = int(0.1 / 0.04)  # 2

    # 使用 `coarsen()` 进行保守降尺度
    ds_coarsened = ds.coarsen(south_north=lat_factor, west_east=lon_factor, boundary="trim").mean()

    # 生成新的经纬度坐标
    new_lat = ds.XLAT.values[::lat_factor, ::lon_factor]
    new_lon = ds.XLONG.values[::lat_factor, ::lon_factor]

    # 更新降尺度后的坐标
    ds_coarsened = ds_coarsened.assign_coords({
        "XLAT": (["south_north", "west_east"], new_lat),
        "XLONG": (["south_north", "west_east"], new_lon)
    })

    return ds_coarsened

# 执行降尺度
ds_downscaled = conservative_downscale_04_to_01(ds)
```

```python
import numpy as np
import xarray as xr
from scipy.spatial import cKDTree

def conservative_remap_aligned(ds, target_resolution=0.1):
    """
    保守重映射到严格对齐的网格（0.1° 间隔）。
    例如：
    - lon: -125.0, -124.9, -124.8, ...
    - lat: 50.0, 49.9, 49.8, ...

    参数:
    - ds: xarray.Dataset, 包含 PREC_ACC_NC (Time, south_north, west_east)
    - target_resolution: 目标分辨率（默认 0.1°）

    返回:
    - ds_remapped: xarray.Dataset, 经过优化保守重映射的新数据集
    """

    # 1️⃣ 获取原始网格
    lons = ds.XLONG.values
    lats = ds.XLAT.values
    data = ds.PREC_ACC_NC[0].values  # 取第一个时间步

    lat_min, lat_max = lats.min(), lats.max()
    lon_min, lon_max = lons.min(), lons.max()

    # 2️⃣ 生成对齐的目标网格
    new_lons = np.arange(np.floor(lon_min * 10) / 10, np.ceil(lon_max * 10) / 10, target_resolution)
    new_lats = np.arange(np.ceil(lat_max * 10) / 10, np.floor(lat_min * 10) / 10, -target_resolution)  # 纬度倒序

    new_lon_grid, new_lat_grid = np.meshgrid(new_lons, new_lats)

    # 3️⃣ 创建 KDTree 加速搜索
    orig_points = np.vstack([lons.flatten(), lats.flatten()]).T
    tree = cKDTree(orig_points)

    # 4️⃣ 计算新网格的加权值
    remapped_data = np.zeros((len(new_lats), len(new_lons)))
    
    for i in range(len(new_lats)):
        for j in range(len(new_lons)):
            # 获取目标网格中心点
            target_point = np.array([new_lons[j], new_lats[i]])

            # 找到最近的 4 个原始网格点（最近邻搜索）
            dists, idxs = tree.query(target_point, k=4)
            
            # 计算权重（使用 1/distance 加权）
            weights = 1 / (dists + 1e-6)  # 避免除零
            weights /= weights.sum()  # 归一化权重

            # 计算加权平均
            remapped_data[i, j] = np.sum(weights * data.flatten()[idxs])

    # 5️⃣ 确保总量守恒
    orig_total = np.nansum(data)
    new_total = np.nansum(remapped_data)
    correction_factor = orig_total / new_total if new_total > 0 else 1
    remapped_data *= correction_factor

    # 6️⃣ 创建新的 xarray 数据集
    ds_remapped = xr.Dataset(
        {
            "PREC_ACC_NC": (["lat", "lon"], remapped_data)
        },
        coords={
            "lat": new_lats,
            "lon": new_lons
        }
    )

    return ds_remapped

```
