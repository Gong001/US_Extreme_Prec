```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load US shapefile
US = gpd.read_file('../../tl_2019_us_state/tl_2019_us_state.shp')

# Dictionary of regions with states
regions_dict = {
    'NE': ['CT', 'DE', 'ME', 'MD', 'MA', 'NH', 'NJ', 'NY', 'PA', 'RI', 'VT', 'WV'],
    'Midwest': ['IA', 'MI', 'MN', 'WI', 'IL', 'IN', 'MO', 'OH'],
    'SE': ['AL', 'FL', 'GA', 'NC', 'SC', 'VA', 'TN', 'KY', 'AR', 'LA', 'MS'],
    'NGP': ['MT', 'NE', 'ND', 'SD', 'WY'],
    'SGP': ['KS', 'OK', 'TX'],
    'SW': ['AZ', 'CO', 'NM', 'UT', 'CA', 'NV'],
    'NW': ['ID', 'OR', 'WA']
}

# Create a GeoDataFrame for the specified regions
regions = {name: US[US['STUSPS'].isin(states)] for name, states in regions_dict.items()}

# Combine all regions into a single GeoDataFrame for plotting
all_regions = gpd.GeoDataFrame(pd.concat(regions.values(), ignore_index=True))

# Color map for the regions
color_map = {
    'NE': '#6a0dad',  # purple
    'Midwest': '#ff0000',  # red
    'SE': '#ffa500',  # orange
    'NGP': '#00ff00',  # lime green
    'SGP': '#006400',  # dark green
    'SW': '#8b0000',  # dark red
    'NW': '#0000ff'   # blue
}

# Assign colors to each region
for region, df in regions.items():
    df['color'] = color_map[region]

# Plotting the map with specified colors and bold boundaries
fig, ax = plt.subplots(figsize=(10, 8))
for region, df in regions.items():
    df.plot(ax=ax, edgecolor='black', linewidth=1.5, color=df['color'].iloc[0], label=region)

ax.set_title('US Regions Map')
ax.set_axis_off()
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Load US shapefile
US = gpd.read_file('../../tl_2019_us_state/tl_2019_us_state.shp')

# Dictionary of regions with states
regions_dict = {
    'NE': ['CT', 'DE', 'ME', 'MD', 'MA', 'NH', 'NJ', 'NY', 'PA', 'RI', 'VT', 'WV'],
    'Midwest': ['IA', 'MI', 'MN', 'WI', 'IL', 'IN', 'MO', 'OH'],
    'SE': ['AL', 'FL', 'GA', 'NC', 'SC', 'VA', 'TN', 'KY', 'AR', 'LA', 'MS'],
    'NGP': ['MT', 'NE', 'ND', 'SD', 'WY'],
    'SGP': ['KS', 'OK', 'TX'],
    'SW': ['AZ', 'CO', 'NM', 'UT', 'CA', 'NV'],
    'NW': ['ID', 'OR', 'WA']
}

# Create a GeoDataFrame for the specified regions
regions = {name: US[US['STUSPS'].isin(states)] for name, states in regions_dict.items()}

# Combine all regions into a single GeoDataFrame for plotting
all_regions = gpd.GeoDataFrame(pd.concat(regions.values(), ignore_index=True))
```

```python
# Set up the map with Albers Equal Area projection
fig, ax = plt.subplots(figsize=(10, 4),dpi=300,subplot_kw={'projection': ccrs.AlbersEqualArea(central_longitude=-96, central_latitude=37.5)})
ax.set_extent([-125, -66, 24, 50], crs=ccrs.PlateCarree())


# Plot each region with its respective color
for region, df in regions.items():
    ax.add_geometries(df['geometry'], crs=ccrs.PlateCarree(),
                      facecolor='#ffffff', edgecolor='black', linewidth=1, label=region)
plt.savefig('figsave/US_map.png')
plt.show()
```
